```python
import os
import glob
import pandas as pd
from config import DATA_ANALYSIS_PATH
```

```python
os.chdir('/mnt/nas/Forest/MG-RAST_Dataset_Analysis')
id_list = os.listdir(DATA_ANALYSIS_PATH)
```

```python
headers = ['Strain','Gene','PercentIdentity','FoldCoverage']
df = pd.DataFrame(columns=headers)
```

```python
for id in id_list:
    try:
        os.chdir(DATA_ANALYSIS_PATH + id + '/reports')
        file = glob.glob('*.csv')
        temp_df = pd.read_csv(file[0])
        df = df.append(temp_df)
    except FileNotFoundError:
        print('Analysis not complete yet.')
        break

df.Gene.unique()
```
